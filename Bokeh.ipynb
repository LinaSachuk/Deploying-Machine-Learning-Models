```python
from bokeh.plotting import figure
import numpy as np

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)


plot = figure(title="Sine Function")
plot.xaxis.axis_label='x'
plot.yaxis.axis_label='amplitude'


plot.line(x, y, 
  line_color='blue',
  line_width=3,
  legend='sin(x)')

plot.circle(x, 2*y, 
  fill_color='red',
  line_color='black',
  fill_alpha=0.2,
  size=10,
  legend='2sin(x)')

#line_dash is an aribrary length list of lengths
#alternating in [color, blank, color, ...]
plot.line(x, np.sin(2*x),
  line_color='green',
  line_dash=[10,10,2,5], 
  line_width=2,
  legend='sin(2x)')

show(plot)
```

```python
import pandas as pd
import numpy as np

# build some distributions
mu, sigma = 0, 0.5
normal = pd.DataFrame({'value': np.random.normal(mu, sigma, 1000), 'type': 'normal'})
lognormal = pd.DataFrame({'value': np.random.lognormal(mu, sigma, 1000), 'type': 'lognormal'})

# create a pandas data frame
df = pd.concat([normal, lognormal])
df[995:1005]
```

```python
from bkcharts import Histogram
hist = Histogram(df, values='value', color='type', bins=50, legend=True)
show(hist)
```
